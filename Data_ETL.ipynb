```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
cancer_death_by_state = pd.read_csv('https://query.data.world/s/5tsppwl3vgs5kqytojxi6keodmlpuu')
cancer_incidence_by_state = pd.read_csv('https://query.data.world/s/b6lyivz7vfyhr2jbr3rdfpqaz5iswf')
home_health_care_quality_by_state= pd.read_csv('https://query.data.world/s/sgwtk73pzldvlp4ppjpx4e4wx6k4ox')
```

```python
home_health_cols = ["State", "Quality of Patient Care Star Rating","How often patients receiving home health care needed urgent, unplanned care in the ER without being admitted"]
home_health_transformed= home_health_care_quality_by_state[home_health_cols].copy()
home_health_transformed = home_health_transformed.rename(columns={"Quality of Patient Care Star Rating": "Star Rating",
                                                          "How often patients receiving home health care needed urgent, unplanned care in the ER without being admitted": "ER Frequency"})
home_health_transformed.head()
```

```python
cancer_death_incidences = pd.merge(cancer_death_by_state, cancer_incidence_by_state, on="State")
cancer_death_incidences_df = cancer_death_incidences.rename(columns={"Range_x":"Death Range", "Rate_x":"Death Rate","Range_y":"Incidence Range", "Rate_y":"Incidence Rate"})
cancer_death_incidences_df.head()
```
